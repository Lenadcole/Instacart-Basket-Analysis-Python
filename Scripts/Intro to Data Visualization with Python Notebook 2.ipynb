#### 3. Create charts

```python
# Check 'order_hour'of_day' column value counts

df_cust_ords_prods['order_hour_of_day'].value_counts()
```

```python
# Create histogram of 'order_hour_of_day' column and save within hist object

hist = df_cust_ords_prods['order_hour_of_day'].plot.hist(bins = 24)
```

The histogram shows frequency of sales across the day in one hour time slots. The higher the frequency, the higher the number of sales at that hour of the day. 
This histogram shows that sales peak at 10am, with the busiest time of day being between 9am and 5pm. The quietest time for sales is early morning between 2am and 5am.

```python
df_cust_ords_prods['loyalty_flag'].value_counts()
```

```python
# Create a bar chart from the 'loyalty_flag' column
```

```python
bar = df_cust_ords_prods['loyalty_flag'].value_counts().plot.bar()
```

It is clear that there are more orders made by Regular customers (those who have made between 10 and 40 orders) than Loyal or New customers.

```python
# Create a sample dataframe in order to check whether there is a difference in expenditure depending on the hour of day

np.random.seed(8)
dev = np.random.rand(len(df_cust_ords_prods)) <= 0.7
```

```python
# Store 70% of the sample in dataframe big

big = df_cust_ords_prods[dev]
```

```python
# Store 30% of the sample in dataframe small

small = df_cust_ords_prods[~dev]
```

```python
# Check both samples add up to the same length as the entire dataframe
```

```python
len(df_cust_ords_prods)
```

```python
len(big) + len(small)
```

```python
df_cust_ords_prods_small = small[['order_hour_of_day','prices']]
```

```python
# Create line chart to check whether there is a difference in expenditure depending on the hour of day
```

```python
line = sns.lineplot(data = df_cust_ords_prods_small, x = 'order_hour_of_day', y = 'prices')
```

Although the line chart shows some peaks and troughs, the scale on the y-axis (between \\$7.75 and \\$7.95) and shows us that there is very little difference in expenditure depending on the hour of the day.

#### 4. Conduct exploratory analysis of customer demographics

```python
df_cust_ords_prods_small_2 = small[['age','number_of_dependants']]
```

```python
# Create line chart to explore the connections between age and number of dependants
```

```python
line_2 = sns.lineplot(data = df_cust_ords_prods_small_2, x = 'age', y = 'number_of_dependants')
```

The line chart shows that there is no connection between age and family situation - it is clear that the number of dependants a person has is based on factors other than age.

```python
# Create scatterplot to explore connection between age and spending power
```

```python
scatterplot = sns.scatterplot(x = 'age', y = 'income', data = df_cust_ords_prods)
```

The scatterplot shows that there is a connection between age and spending power. The data shows that, while the majority of Instacart customers have an income of below \\$200,000, there is a not-insignificant number of customers whose income exceeds that and reaches (in some cases) around \\$400,000 for those aged 40 and below. The number of customers with an income of over \\$200,000 increases significantly above the age of 40, with income rising to over \\$600,000 in some cases.
