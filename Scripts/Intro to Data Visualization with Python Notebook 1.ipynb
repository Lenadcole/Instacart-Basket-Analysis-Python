#### 4. Consistency and data quality checks

```python
# Check column data types

df_cust.dtypes
```

```python
# Check for mixed data types

for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].map(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

```python
# Covert first_name column data type from mixed to string

df_cust['first_name'] = df_cust['first_name'].astype('str')
```

```python
# Check for missing values

df_cust.isnull().sum()
```

When I was looking into outliers earlier, I noticed that a customer with a user id of 98580 had NaN in the first_name column, denoting a missing value. However, after changing datatype for that column and running the code to see if there were any missing values, a value of 0 was returned. I decided to see what the missing value had become.

```python
# Check what missing values have become

df_cust.loc[df_cust['user_id'] == 98580]
```

I discovered that Python had changed NaN to the word nan! While this could be a first name, it would be better to change it to a dash to denote that there is no first name associated with this customer. I would not do this for a numeric value or important information - in those cases I would create a flag or impute the value with the mean or median of the column. But in this case I don't believe the information missing will have any effect on the rest of the customer's data as we have their user id and surname.

```python
# Check number of missing values

df_cust.loc[df_cust['first_name'] == 'nan']
```

```python
# Change 'nan' to '-' to denote no first name data

df_cust['first_name'] = df_cust['first_name'].replace(['nan'], '-')
```

```python
# Check changes

df_cust.loc[df_cust['first_name'] == '-']
```

The change has returned the same number of rows as the original number of missing values.
