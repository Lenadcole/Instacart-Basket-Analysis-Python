#### 3. Creating new variables - price_label and busiest_day

```python
# Create new variables
```

```python
# Define a function

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid range product'
    elif row['prices'] > 15:
          return 'High range product'
    else: return np.nan
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

#### 4. Create busiest_days variable

#### 5. Create busiest_period_of_day variable

```python
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

Looking at the value counts for the order_hour_of_day column, I think the best way of dividing the periods of time into 'most orders', 'average orders' and 'fewest orders' would be as follows:

Most orders - orders above or equal to 1,000,000. The applicable hours are between 0900 and 1700 hours.

Average orders - orders below 1,000,000 but above or equal to 400,000. The applicable hours are 0700, 0800 and between 1800 and 2300 hours.

Fewest orders - orders below 400,000. The applicable hours are between 0000 and 0600 hours

```python
# Create busiest_period_of_day variable

busiest_period_of_day_result = []

for value in df_ords_prods_merge['order_hour_of_day']:
  if value in range (9,18):
    busiest_period_of_day_result.append('Most orders')
  elif value in range (0,7):
    busiest_period_of_day_result.append('Fewest orders')
  else:
    busiest_period_of_day_result.append('Average orders')
```

```python
busiest_period_of_day_result
```

```python
df_ords_prods_merge['busiest_period_of_day'] = busiest_period_of_day_result
```

```python
# Check frequency for busiest_period_of_day column

df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
